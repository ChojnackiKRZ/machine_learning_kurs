Dyskretyzacja zmiennej ciągłej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
```

```python
'''
Ilosc klas wyliczamy pierwiatkujac liczebnosc proby
Dlugosc klas wyliczamy dzielac rozstep przez ilosc klas
'''
ilosc_klas = int (round (len(df)**(1/2), 0))
rozstep = max(df['height']) - min(df['height'])
dlugosc_klas = rozstep/ilosc_klas
df['height_cat'] = pd.cut(x=df.height, bins=3)
```

```python
binnedPopulation = pd.cut(df['height'], ilosc_klas)
tabela_czestosci = binnedPopulation.value_counts()
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195))
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195), labels=['small', 'medium', 'high'])
df
```

```python
pd.get_dummies(df, drop_first=True, prefix='height')
```
