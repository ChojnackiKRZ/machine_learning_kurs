###PODSTAWOWE DANE DOT. ZBIORU

```python
import sklearn
import numpy as np
import pandas as pd

```

```python
sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df
```

```python
df.describe().T
```

```python
df.describe(include = ['category']).T
```

###BINARNE ENCODOWANIE TARGETU

LABEL ENCODER STOSUJEMY DO TARGETU! DO LABELEK STOSUJEMY INNE NARZĘDZIA

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])

df
```

```python
le.inverse_transform(df['bought'])
```

###BINARNE KODOWANIE LABELEK

```python
'''
parametr sparse pozwala nam zwrocic macierz rzadką (True) lub zwykla (False). 
Macierz rzadka przechwouje tylko pozycje 1, dlatego nie możemy jej podejrzec.
Oszednosc pamieci
'''

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

###STANDARYZACJA: StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

###PEŁNE PRZYGOTOWANIE DANYCH DO MODELU - WSZYSTKIE KROKI JESZCZE RAZ

```python
df = df_raw.copy()
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
scaler = StandardScaler()

df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, drop_first=True)
df
```
